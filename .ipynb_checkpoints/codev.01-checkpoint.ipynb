```python
#Practice 2
s = Service("/Users/Theo/Dev/scrappingProject/chromedriver")
driver = webdriver.Chrome(service=s)

driver.get("https://www.memoiredeshommes.sga.defense.gouv.fr/fr/arkotheque/client/mdh/opex_1905_1962/index.php")
results = []
content = driver.page_source
page = BeautifulSoup(content)
driver.quit()

for i in page.findAll(attr="MDHFormChamp champ_formulaire champ_un_quart"):
    name = i.find("option value=27")
    if name not in results:
        results.append(name.text)
print(results)
        
```

```python
#Practice 3 : utiliser le moteur de recherche du SHD avec selenium, on essaye juste de chercher un nom 
#The info for the box we need to fill in:
#r_c_nom


def search(nom):
    url = "https://www.memoiredeshommes.sga.defense.gouv.fr/fr/arkotheque/client/mdh/opex_1905_1962/index.php"
    browser = webdriver.Chrome(service=s)
    browser.get(url)
    search_box = browser.find_element_by_name("r_c_nom")
    search_box.send_keys(nom)
    search_box.send_keys(Keys.RETURN)
    
    
```

```python
search("Abd")
```
